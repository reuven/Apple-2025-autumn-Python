# Exercise: Taxi data

1. Load the file, `taxi.csv`, into a data frame.
2. In how many trips was the `trip_distance` 0? In these trips, how much did people pay (`total_amount`), on average?
3. In how many trips were there 0 passengers (`passenger_count`)? How much did they pay (`total_amount`), and how far did they drive (`trip_distance`)?

Download from : https://files.lerner.co.il/data-science-exercise-files.zip (3rd link at https://files.lerner.co.il)

```python
df = pd.read_csv('taxi.csv',
                usecols=['trip_distance', 'passenger_count', 'total_amount'])
df
```

```python
# 2. In how many trips was the `trip_distance` 0? In these trips, how much did people 
# pay (`total_amount`), on average?

(
    df['total_amount']
    .loc[df['trip_distance'] == 0]
    .mean()
)
```

```python
# 3. In how many trips were there 0 passengers (`passenger_count`)? How much did they pay (`total_amount`), 
# and how far did they drive (`trip_distance`)?


(
    df[['total_amount', 'trip_distance']]
    .loc[df['passenger_count'] == 0]
    .mean()
)
```
