# Some attributes

- `.values` -- returns the NumPy array used for storage
- `.loc` -- uses `[]`, returns one or more values from our series based on the index
- `.iloc` -- uses `[]`, return one or more values from our series based on the *position*

```python
s.iloc[-1]
```

```python
s.loc[0]
```

```python
s.values
```

```python
s.values[0]
```

```python
s.loc[2:5]   
```

```python
s.iloc[2:5]
```

```python
# fandcy indexing
s.loc[[2, 4, 1]]
```

```python
s.iloc[[2, 4, 1]]
```
