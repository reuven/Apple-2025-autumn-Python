```python
np.random.seed(0)
s = Series(np.random.randint(0, 1000, 15), 
          index=list('abcde')*3)
s
```

```python
s.loc[(s >= s.mean() - s.std())  & (s <= s.mean() + s.std())] 
```

```python
def non_outlier(a_series):
    return ((a_series >= a_series.mean() - a_series.std())  & 
            (a_series <= a_series.mean() + a_series.std()))

s.loc[non_outlier]  # .loc calls non_outlier(s) --> function returns a boolean series
```

```python
s.loc[lambda s_: (s_ >= s_.mean() - s_.std()) & (s_ <= s_.mean() + s_.std())] 
```

```python
(
    s
    .loc[lambda s_: s_ >= s_.mean() - s_.std()]  # keep only those above mean-std
    .loc[lambda s_: s_ <= s_.mean() + s_.std()]  # keep only those below mean+std
    .loc[lambda s_: s_ % 2 == 0]   # keep only even numbers
)
```

```python
(
    s
    .loc[s >= s.mean() - s.std()]  # keep only those above mean-std
    .loc[s <= s.mean() + s.std()]  # keep only those below mean+std
    .loc[s: s % 2 == 0]   # keep only even numbers
)
```

```python
(
    s
    .loc[lambda s_: s_ >= s_.mean() - s_.std()]  # keep only those above mean-std
    .loc[lambda s_: s_ <= s_.mean() + s_.std()]  # keep only those below mean+std
    .loc[lambda s_: s_.isin([754, 804])]   # if s is either 754 or 804, return True
)
```
