Get the files from https://files.lerner.co.il, the final link there, https://files.lerner.co.il/pandas-workout-data.zip

We'll use the olympic CSV file from there (or from this repo)

1. Create a data frame from this data. Use four columns in the index: `Year`, `Season`, `Sport`, `Event`.
2. Find the mean age for summer games.
3. Which team won the most medals in archery?
4. How tall was the tallest-ever tennis player?

```python
df = pd.read_csv('olympic_athlete_events.csv',
                index_col=['Year', 'Season', 'Sport', 'Event'])
df
```

```python
# Find the mean age for summer games.

(
    df
    .xs('Summer', level='Season')
    ['Age']
    .mean()
)
```

```python
# Which team won the most medals in archery?

(
    df
    .xs('Archery', level='Sport')
    [['Team', 'Medal']]
    .dropna()
    ['Team']
    .value_counts()
)
```

```python

(
    df
    .xs('Archery', level='Sport')
    .dropna(subset='Medal')
    ['Team']
    .value_counts()
)
```
