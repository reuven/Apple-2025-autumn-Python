```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('taxi.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# I want all rows with passenger_count > 1

(
    df
    .loc[df['passenger_count'] == 1]
)
```

```python

(
    df
    .set_index('passenger_count')
    .loc[1]
)
```

```python

(
    df
    .set_index('passenger_count')
    .loc[[2, 3]]
)
```

```python
# I want all rows with 2, 3, or 4 passengers

(
    df
    .set_index('passenger_count')
    .loc[[2,3,4]]    # fancy indexing
)
```

```python
(
    df
    .set_index('passenger_count')
    .sort_index()
    .loc[2:4]
    ['trip_distance']   # select the trip_distance column
    .mean()
)
```

```python
(
    df
    .set_index('passenger_count')
    .sort_index()
    .loc[
        2:4,    # row selector 
        'trip_distance'  # column selector
        ]
    .mean()
)
```

```python
(
    df
    .set_index('passenger_count')
    .sort_index()
    .loc[2:4]
    .reset_index()
    [['trip_distance', 'passenger_count']]
   .mean()
)
```
