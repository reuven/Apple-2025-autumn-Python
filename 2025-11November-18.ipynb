# Exercise: Olympic data

Get the files from https://files.lerner.co.il, the final link there, https://files.lerner.co.il/pandas-workout-data.zip

We'll use the olympic CSV file from there (or from this repo)

1. Create a data frame from this data. Use four columns in the index: `Year`, `Season`, `Sport`, `Event`.
2. Find the mean age for summer games.
3. Which team won the most medals in archery?
4. How tall was the tallest-ever tennis player?

```python
df = pd.read_csv('olympic_athlete_events.csv',
                index_col=['Year', 'Season', 'Sport', 'Event'])
df
```

```python
# Find the mean age for summer games.

(
    df
    .xs('Summer', level='Season')
    ['Age']
    .mean()
)
```

```python
# Which team won the most medals in archery?

(
    df
    .xs('Archery', level='Sport')   # keep only rows where 'Sport' in our multi-index is 'Archery'
    [['Team', 'Medal']]
    .dropna()    # remove any row with NaN anywhere
    ['Team']
    .value_counts()
)
```

```python
(
    df
    .xs('Archery', level='Sport')   # keep only where sport is archery
    .dropna(subset='Medal')         # only remove rows where Medal is NaN
    ['Team']                         # keep the 'Team' column
    .value_counts()                
)
```

```python
# How tall was the tallest-ever tennis player?

(
    df
    .xs('Tennis', level='Sport')  # keep only rows where Sport == 'Tennis'
    .set_index('Team')
    ['Height']                     # keep only 'Height' column
    .agg(['max', 'idxmax'])
)
```

# Exercise: Olympic grouping

1. Create a new data frame from Olympic data.
2. How many medals did each team receive, starting in 1990? What 10 countries got the  most?
3. Find the 10 sports whose athletes are, on average, the tallest.
4. From 1990, find the mean height + weight for each combination of sport + team.

```python
df = pd.read_csv('olympic_athlete_events.csv')
df
```

```python
%%timeit

# 2. How many medals did each team receive, starting in 1990? What 10 countries got the  most?
(
    df
    .set_index('Year')
    .sort_index()
    .loc[1990:]
    .groupby('Team')['Medal'].count()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
%%timeit

(
    df
    .set_index('Year')
    .sort_index()
    .loc[1990:]
    .groupby('Team')['Medal'].count()
    .nlargest(10)
)
```

```python
# 3. Find the 10 sports whose athletes are, on average, the tallest.

(
    df
    .groupby('Sport')['Height'].mean()
    .nlargest(10)
)
```

```python
# 4. From 1990, find the mean height + weight for each combination of sport + team.

(
    df
    .loc[ df['Year'] >= 1990 ]
    .groupby(['Team', 'Sport'])[['Height', 'Weight']].mean()
)
```
